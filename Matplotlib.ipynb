```python
# plotting a simple function

price = [48000 ,54500 ,57000, 49000, 475000, 40000, 100000]
year = [2015, 2016, 2017, 2018, 2019, 2020, 2021]

# syntax --> plt.plot(x-axis, y-axis)
plt.plot(year, price)

# generally the limit is introduced when there's a outlier vlaue in the list of vlaues

plt.xlim(2015,2021)# I want to see the graph between 2018-2021
plt.ylim(50000,100000)
plt.show()
```

```python
# BAR CHART

# Simple bar chart

children = [20, 5, 10, 2, 15]
colors = ['Red', 'Green', 'Blue', 'Yellow', 'Violet']
plt.bar(colors, children)
plt.title('Favourite Color')
plt.xlabel('Colors')
plt.ylabel('Children')
```

```python
# Horizontal Bar Chart
# This is done when there are more categories on the x-axis as compared to the y-axis
plt.barh(colors, children, color='Orange')
plt.title('Favourite Color')
plt.xlabel('Colors')
plt.ylabel('Children')
```

```python
# Multiple Bar Charts

df = pd.read_csv('/content/batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'], df['2015'])
```

```python
# If we plot multiple graphs together , it will overlap
plt.bar(df['batsman'], df['2015'])
plt.bar(df['batsman'], df['2016'])
plt.bar(df['batsman'], df['2017'])
```

```python
# Width of the bars
plt.bar(df['batsman'], df['2015'], width=0.2)
```

```python
# Replacing X-Axis with a set of numbers

np.arange(df.shape[0])
```

```python
# Plotting Multiple Graphs together
# the value of displacement shd be equal to the value of width
# then only a perfect graph would be plotted
plt.bar(np.arange(df.shape[0])-0.2, df['2015'], width=0.2, color='g')
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2, color='c')
plt.bar(np.arange(df.shape[0])+0.2, df['2017'], width=0.2, color='r')

# Now Again Replacing the X-Axis ticks with the Original Names
# Syntax --> plt.xticks(current_x-axis, to_be_x-axis)

plt.xticks(np.arange(df.shape[0]), df['batsman'])
plt.title("IPL Season Record")
plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.show()
```

```python
# Problem
# when there's too big names of the x-axis then those names overlap to each other and the readability becomes 0
# this also happens when there is a greater number of categories

children = [20, 5, 10, 2, 15]
colors = ['Red Red Red Red', 'Green Green Green Green', 'Blue Blue Blue', 'Yellow Yellow Yellow', 'Violet Violet Violet']
plt.bar(colors, children)
plt.title('Favourite Color')
plt.xlabel('Colors')
plt.ylabel('Children')
```

```python
# TO SOLVE THE ABOVE ISSUE
# We use a function plt.xticks(rotation='vertical')
children = [20, 5, 10, 2, 15]
colors = ['Red Red Red Red', 'Green Green Green Green', 'Blue Blue Blue', 'Yellow Yellow Yellow', 'Violet Violet Violet']
plt.bar(colors, children)
plt.title('Favourite Color')
plt.xlabel('Colors')
plt.ylabel('Children')
plt.xticks(rotation='vertical')
plt.show()
```

```python
# Stacked Bar Chart

# to avoid overlapping of the bars and to stack them on top of each other without overlapping
# we use (bottom= (value of current x-axis + values of all the above x-axis values too))

plt.bar(df['batsman'], df['2015'], label='2015')
plt.bar(df['batsman'], df['2016'], bottom=df['2015'], label='2016')
plt.bar(df['batsman'], df['2017'], bottom=(df['2016']+df['2015']), label='2017')
plt.legend()
plt.show()
```

```python
# Logarithmic Scale

arr = np.load('/content/big-array.npy')
arr.shape # contains 12k items in it (a very big array)
plt.hist(arr, bins=[10,20,30,40,50,60,70])
plt.show()
```

```python
# Here in the above output , the bins of (0-10) is not visible
# even if there are some values in that range
# this happened due to the presence of a very large number replacing the smaller
# to fix this problem, we use logarithmic scale

plt.hist(arr, bins=[10,20,30,40,50,60,70], log=True)
plt.show()
```

```python
# on dataset
df = pd.read_csv('/content/gayle-175.csv')
df
```

```python
# Here the autopct='%0.1f%%' is a necessary code to calculate and print the percentage

plt.pie(df['batsman_runs'], labels=df['batsman'], autopct = '%0.1f%%')
plt.show()
```

```python
# Percentage and colors

plt.pie(df['batsman_runs'], labels=df['batsman'], autopct = '%0.1f%%', colors=['orange', 'purple', 'yellow', 'pink', 'cyan', 'red'])
plt.show()
```

```python
# Explode Shadow(ek block ko bahar nikalna)

plt.pie(df['batsman_runs'], labels=df['batsman'], autopct = '%0.1f%%', explode=[0,0,0,0.6,0,0], shadow=True)
plt.show()
```
